```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_excel('Online Retail.xlsx')
```

```python
France_df = df[df['Country']=='France']
```

```python
France_df.reset_index(drop=True, inplace=True)
```

```python
France_df
```

```python
pivot_France_df = pd.pivot_table(data = France_df, index = 'InvoiceNo', columns = 'Description', values='Quantity').fillna(0)
pivot_France_df
```

```python
encoded_France_df = pivot_France_df.map(lambda x:1 if x>0 else 0)
```

```python
frequent_dataset = apriori(encoded_France_df, min_support=0.05, verbose=1, use_colnames=True)
```

```python
frequent_dataset
```

```python
best_association = association_rules(frequent_dataset, metric='confidence', min_threshold=0.7, num_itemsets=len(frequent_dataset))
```

```python
best_association[['antecedents', 'consequents', 'support', 'confidence']]
```
