```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
from sklearn.feature_selection import SelectFromModel, RFE
```

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data=data.data, columns=data.feature_names)
```

```python
df['Target'] = data.target
```

```python
x = df.drop(labels='Target', axis=1)
```

```python
y = df[['Target']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.75)
```

```python
df.shape
```

```python
df.dtypes
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
selectFromModel = SelectFromModel(estimator=RandomForestClassifier(n_estimators=50, max_depth=6))
```

```python
selectFromModel.fit(x_train, y_train)
```

```python
selectFromModel.get_support()
```

```python
x.columns[selectFromModel.get_support()]
```

```python
x_train_transformed = selectFromModel.transform(x_train)
```

```python
x_train_transformed.shape
```

```python
x_test_transformed = selectFromModel.transform(x_test)
```

```python
x_test_transformed.shape
```

```python
model.fit(x_train_transformed, y_train)
```

```python
y_pred = model.predict(x_test_transformed)
accuracy_score(y_test, y_pred)
```

# RFE

```python
rfe_model = RFE(estimator=RandomForestClassifier(n_estimators=50, max_depth=6), n_features_to_select=15)
```

```python
rfe_model.fit(x_train, y_train)
```

```python
rfe_model.fit(x_train, y_train)
```

```python
x.columns[rfe_model.get_support()]
```

```python
x_train_transformed = rfe_model.transform(x_train)
x_test_transformed  = rfe_model.transform(x_test)
```

```python
model.fit(x_train_transformed, y_train)
```

```python
y_pred = model.predict(x_test_transformed)
accuracy_score(y_test, y_pred)
```

```python
for i in range(1, 31):
    rfe_model = RFE(estimator=RandomForestClassifier(n_estimators=50, max_depth=6), n_features_to_select=i)
    rfe_model.fit(x_train, y_train)
    x_train_transformed = rfe_model.transform(x_train)
    x_test_transformed  = rfe_model.transform(x_test)
    model.fit(x_train_transformed, y_train)
    y_pred = model.predict(x_test_transformed)
    acc = accuracy_score(y_test, y_pred)

    print(f"Accuracy score for {i} is {acc}")
```
